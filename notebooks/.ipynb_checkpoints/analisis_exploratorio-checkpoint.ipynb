```python
# Importar librer√≠as
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Configuraci√≥n de visualizaci√≥n
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
%matplotlib inline

# Configurar tama√±o de gr√°ficos
plt.rcParams['figure.figsize'] = (12, 6)

print("‚úÖ Librer√≠as importadas correctamente")
```

```python
# Cargar datos
df = pd.read_csv('../data/Sample - Superstore.csv', encoding='latin-1')

print("üìä Dataset cargado correctamente")
print(f"Dimensiones: {df.shape[0]} filas x {df.shape[1]} columnas")
print("\nüîç Primeras filas:")
df.head()
```

```python
# Informaci√≥n general del dataset
print("=" * 50)
print("INFORMACI√ìN GENERAL DEL DATASET")
print("=" * 50)
df.info()

print("\n" + "=" * 50)
print("ESTAD√çSTICAS DESCRIPTIVAS")
print("=" * 50)
df.describe()
```

```python
# Verificar valores nulos
print("üîç Valores nulos por columna:")
print(df.isnull().sum())

# Verificar duplicados
duplicados = df.duplicated().sum()
print(f"\nüìã Registros duplicados: {duplicados}")

# Convertir columna de fecha a formato datetime
if 'Order Date' in df.columns:
    df['Order Date'] = pd.to_datetime(df['Order Date'])
    df['Ship Date'] = pd.to_datetime(df['Ship Date'])
    
    # Crear columnas adicionales √∫tiles
    df['Year'] = df['Order Date'].dt.year
    df['Month'] = df['Order Date'].dt.month
    df['Month_Name'] = df['Order Date'].dt.month_name()
    
print("\n‚úÖ Limpieza completada")
print(f"Dataset final: {df.shape[0]} registros")
```

```python
# M√©tricas principales
ventas_totales = df['Sales'].sum()
ganancia_total = df['Profit'].sum()
cantidad_pedidos = df['Order ID'].nunique()
margen_promedio = (df['Profit'].sum() / df['Sales'].sum()) * 100

print("=" * 50)
print("üìä M√âTRICAS CLAVE DEL NEGOCIO")
print("=" * 50)
print(f"üí∞ Ventas Totales: ${ventas_totales:,.2f}")
print(f"üìà Ganancia Total: ${ganancia_total:,.2f}")
print(f"üõí N√∫mero de Pedidos: {cantidad_pedidos:,}")
print(f"üìä Margen de Ganancia: {margen_promedio:.2f}%")
```

```python
# An√°lisis por categor√≠a
ventas_categoria = df.groupby('Category').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Order ID': 'nunique'
}).round(2)

ventas_categoria.columns = ['Ventas', 'Ganancia', 'Pedidos']
ventas_categoria = ventas_categoria.sort_values('Ventas', ascending=False)

print("\nüì¶ VENTAS POR CATEGOR√çA:")
print(ventas_categoria)

# Visualizaci√≥n
plt.figure(figsize=(10, 6))
ventas_categoria['Ventas'].plot(kind='bar', color=['#2ecc71', '#3498db', '#e74c3c'])
plt.title('Ventas Totales por Categor√≠a', fontsize=16, fontweight='bold')
plt.xlabel('Categor√≠a', fontsize=12)
plt.ylabel('Ventas ($)', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)

# Guardar gr√°fico
plt.tight_layout()
plt.savefig('../resultados/ventas_por_categoria.png', dpi=300, bbox_inches='tight')
plt.show()

print("‚úÖ Gr√°fico guardado en 'resultados/ventas_por_categoria.png'")
```

```python
# Ventas mensuales
ventas_mensuales = df.groupby(['Year', 'Month'])['Sales'].sum().reset_index()

plt.figure(figsize=(14, 6))
for year in ventas_mensuales['Year'].unique():
    data_year = ventas_mensuales[ventas_mensuales['Year'] == year]
    plt.plot(data_year['Month'], data_year['Sales'], marker='o', label=f'A√±o {year}', linewidth=2)

plt.title('Tendencia de Ventas Mensuales', fontsize=16, fontweight='bold')
plt.xlabel('Mes', fontsize=12)
plt.ylabel('Ventas ($)', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.xticks(range(1, 13))

plt.tight_layout()
plt.savefig('../resultados/tendencia_ventas.png', dpi=300, bbox_inches='tight')
plt.show()

print("‚úÖ Gr√°fico guardado en 'resultados/tendencia_ventas.png'")
```

```python
# Top productos por ventas
top_productos = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 8))
top_productos.plot(kind='barh', color='steelblue')
plt.title('Top 10 Productos por Ventas', fontsize=16, fontweight='bold')
plt.xlabel('Ventas ($)', fontsize=12)
plt.ylabel('Producto', fontsize=12)
plt.gca().invert_yaxis()
plt.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.savefig('../resultados/top_productos.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nüèÜ TOP 10 PRODUCTOS:")
print(top_productos)
```

```python
# Ventas por regi√≥n
ventas_region = df.groupby('Region').agg({
    'Sales': 'sum',
    'Profit': 'sum'
}).round(2)

plt.figure(figsize=(10, 6))
ventas_region['Sales'].plot(kind='bar', color=['#9b59b6', '#e67e22', '#1abc9c', '#34495e'])
plt.title('Ventas por Regi√≥n', fontsize=16, fontweight='bold')
plt.xlabel('Regi√≥n', fontsize=12)
plt.ylabel('Ventas ($)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('../resultados/ventas_por_region.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nüó∫Ô∏è VENTAS POR REGI√ìN:")
print(ventas_region)
```

```python
# Resumen de insights
print("=" * 60)
print("üí° INSIGHTS PRINCIPALES")
print("=" * 60)

# 1. Categor√≠a m√°s rentable
categoria_top = ventas_categoria.index[0]
porcentaje_categoria = (ventas_categoria.loc[categoria_top, 'Ventas'] / ventas_totales) * 100

print(f"\n1Ô∏è‚É£ CATEGOR√çA L√çDER:")
print(f"   ‚Ä¢ '{categoria_top}' genera ${ventas_categoria.loc[categoria_top, 'Ventas']:,.2f}")
print(f"   ‚Ä¢ Representa el {porcentaje_categoria:.1f}% del total de ventas")

# 2. Producto estrella
producto_top = top_productos.index[0]
ventas_producto_top = top_productos.iloc[0]

print(f"\n2Ô∏è‚É£ PRODUCTO ESTRELLA:")
print(f"   ‚Ä¢ '{producto_top}'")
print(f"   ‚Ä¢ Ventas: ${ventas_producto_top:,.2f}")

# 3. Regi√≥n m√°s fuerte
region_top = ventas_region['Sales'].idxmax()
ventas_region_top = ventas_region.loc[region_top, 'Sales']

print(f"\n3Ô∏è‚É£ REGI√ìN M√ÅS FUERTE:")
print(f"   ‚Ä¢ Regi√≥n '{region_top}'")
print(f"   ‚Ä¢ Ventas: ${ventas_region_top:,.2f}")

# 4. Estacionalidad
mes_mayor_venta = ventas_mensuales.loc[ventas_mensuales['Sales'].idxmax(), 'Month']
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']

print(f"\n4Ô∏è‚É£ ESTACIONALIDAD:")
print(f"   ‚Ä¢ Mes con mayores ventas: {meses[int(mes_mayor_venta)-1]}")
print(f"   ‚Ä¢ Recomendaci√≥n: Aumentar inventario en Q4")

print("\n" + "=" * 60)
```
